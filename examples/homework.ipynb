## Code Introduction

I wrote a simple package that can quickly compute the factor of safety for any slope. I will outline the steps below:

```python
#Import the package; check you have necessary dependencies
from python_slopes.calculations import SlopeStabilityCalculator
```

The package is pretty standard python object oriented programming. There are two callable classes: the calculator and slices.

```python
calculator = SlopeStabilityCalculator()
```

You can add as many soil layers as you would like. We have two in the problem.

```python
calculator.add_soil_layer(unit_weight=125, cohesion=600, friction_angle=25,bottom_elevation= 20,top_elevation=60)
calculator.add_soil_layer(unit_weight=110, cohesion=0, friction_angle=30,bottom_elevation= 0,top_elevation=20)
```

We also have the ability to add a piezometric line:

```python
x_piez = [0.0, 50.0, 90.0, 130.0, 180.0]
y_piez = [40.0, 35.0, 30.0, 20.0, 20.0]
calculator.add_piezeometric_line(x_piez, y_piez)
```

You can add several different slip surface geometries; let us stick with circular slip surface.

```python
slip_center = (100.0, 80.0)
slip_radius = 78.0
calculator.add_slip_circle(center=slip_center, radius=slip_radius)
```

You then can add any sort of irregular ground surface. For now, I am simply plotting the homework problem's ground surface as a piecewise function.

```python
def ground_surface(x):
    if x <= 50: return 60.0
    elif x <= 130: return 60 - (40/80) * (x - 50)
    else: return 20.0

calculator.set_ground_surface(ground_surface)
```

Now you can conduct either OMS Bishops or both analyses with a selected number of slices. Note you have the option of two weight calculations: using the total unit weight of the base material or integrating across all the layers' unit weights. I am choosing the latter for this analysis.

```python
calculator.create_slices(num_slices=100)
results = calculator.analyze(methods=['oms', 'bishop'], use_base_unit_weight=False)
print("Factors of Safety:", results)
```

With 100 slices, I got Factor of Safety values of 1.29 and 1.60, respectively. These are slightly off from my hand calculations because I furthered the accuracy in my program. For pore pressures, I integrate across the entire bottom of each slice (vs. just taking the midpoint).

The plotted results are shown below:

```python
calculator.plot_analysis()
```

Values from the first five slices are also given below:

```python
for i, s in enumerate(calculator.slices[:5]):
    print(f"Slice {i}: Weight={s.weight:.2f}, Pore Pressure={s.pore_pressure:.2f}, Alpha={s.alpha:.2f}. Material={s.material}")
```
